```python
import random
import math
import pandas as pd
```

```python
#Number of outcomes of fair die
fair = [0, 0, 0, 0, 0, 0]
def fairdie(trails):
    for i in range(trails):
        op = random.randint(1, 6)
        fair[op - 1] = fair[op - 1] + 1
        
fairdie(250000)
print("Fair dice outcomes in the order of 1-6:", fair)
```

```python
# To count Number of Outcomes of a Biased Die
weights = (0.15, 0.15, 0.15, 0.15, 0.15, 0.25)

def roll(weights):
    random_roll = random.random()
    sum = 0
    op = 1
    for i in weights:
        sum += i
        if random_roll < sum:
            return op
        op+=1

print(roll(weights))
```
